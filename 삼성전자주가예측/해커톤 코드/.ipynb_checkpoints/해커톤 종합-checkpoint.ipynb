# 지표 설명
## 주가
* SAMSUNG: 삼성 주가
* NDAQ: 나스닥 지수
* HSI	: 항셍 지수
* KOSPI: 코스피
* N225: 니케이225
* SOX	: 필라델피아 반도체 지수

## 채권
* KR_3year_yield(%): 국채 3년물 수익률	
* US_10year_yeild(%): 미국 재무성 채권 10년물 수익률 

## 원자재
* Gold : 금 현물 시세(달러 표기)
* Dubai(BPD) : 두바이유(배럴 당 달러) - 한국석유공사 공시 자료
* Brent(BPD): 브렌트유(배럴 당 달러) - 한국석유공사 공시 자료
* WTI(BPD): 서부 텍사스유(배럴 당 달러) -한국석유공사 공시 자료
    - 한국 석유공사가 국제 원유 현물 가격을 공시하는데 해당 가격은 미국의 Mcgraw Hill사가 운영하는 Platt's에서 전세계 현물 거래 정보를 수집하여 고유 가격 산정 방식에 따라서 적당한 당일 평가 가격을 게시하고 이 자료를 그대로 올리고 있다.-> 산업통상자원부 자료 참조
    - 금 지수는 2,4,6,8,10월 중 해당 날짜에서 가장 가까운 만기의 선물 가격을 고시하고 있음 ->CME(시카고 상품 거래소)

## 환율
* JPY(1YEN): 1엔당 원	
* USD: 1달러 당 원


```python
# 삼성 주가 및 항셍,나스닥, 니케이, 코스피, 필라델피아 반도체 지수
import yfinance as yf
import pandas as pd
import FinanceDataReader as fdr
import pandas_datareader as pdr

# 2018-12-25~2021-12-31까지 지수 호출
stock_index= yf.download(['^KS11','^HSI','^N225','NDAQ','^SOX','005930.KS'],start = '2018-12-25',end='2021-12-31')['Adj Close']
stock_index.rename(columns={'^KS11':'KOSPI','^HSI':'HSI','^N225':'N225','NDAQ':'NDAQ','^SOX':'SOX','005930.KS':'SAMSUNG'},inplace=True)

# 결측치 처리 후 기간 설정(2019-1-1~2021-12-31) 후 출력
stock_index = stock_index.interpolate(method='time').loc['2019-1-1':'2021-12-31',:]
stock_index
```

```python
# 원자재: 금현물, 브렌트, 두바이, 텍사스 선물
## 금가격
Gold = pdr.get_data_yahoo('GC=F', '2019-01-01', '2021-12-31')
Gold_set = Gold.rename(columns={'Adj Close':'Gold($)'})['Gold($)']
Gold_set
```

```python
#20190101~20191231 유가
from datetime import datetime
import time

oil1 = pd.read_csv('oil_20190101_20191231.csv',encoding='cp949')

oil1.drop('Unnamed: 1',axis=1,inplace=True)
oil1['Date']=oil1['Date'].apply(lambda x: x.replace('?',''))
oil1['Date']=oil1['Date'].apply(lambda x: x.replace('월','-'))
oil1['Date']=oil1['Date'].apply(lambda x: x.replace('일',''))
oil1['Date']=oil1['Date'].apply(lambda x: '2019-'+x)
oil1['Date']=pd.to_datetime(oil1['Date'],infer_datetime_format=True)
oil1.set_index('Date',inplace=True)
oil1
# 
```

```python
#20200101~20201231 유가
oil2 = pd.read_csv('oil_20200101_20201231.csv',encoding='cp949')
oil2.drop('Unnamed: 1',axis=1,inplace=True)
oil2['Date']=oil2['Date'].apply(lambda x: x.replace('?',''))
oil2['Date']=oil2['Date'].apply(lambda x: x.replace('월','-'))
oil2['Date']=oil2['Date'].apply(lambda x: x.replace('일',''))
oil2['Date']=oil2['Date'].apply(lambda x: '2020-'+x)
oil2['Date']=pd.to_datetime(oil2['Date'],infer_datetime_format=True)
oil2.set_index('Date',inplace=True)
oil2
```

```python
#20210101~20211231 유가
oil3 = pd.read_csv('oil_20210101_20211231.csv',encoding='cp949')
oil3.drop('Unnamed: 1',axis=1,inplace=True)
oil3['Date']=oil3['Date'].apply(lambda x: x.replace('?',''))
oil3['Date']=oil3['Date'].apply(lambda x: x.replace('월','-'))
oil3['Date']=oil3['Date'].apply(lambda x: x.replace('일',''))
oil3['Date']=oil3['Date'].apply(lambda x: '2021-'+x)
oil3['Date']=pd.to_datetime(oil3['Date'],infer_datetime_format=True)
oil3.set_index('Date',inplace=True)
oil3
```

```python
# 유가 종합
oil = pd.concat([oil1,oil2,oil3])
oil.rename(columns={'Dubai':'Dubai(BPD)','Brent':'Brent(BPD)','WTI':'WTI(BPD)'},inplace=True)
oil.apply(lambda x: x.replace('-',None,inplace=True))
oil.dtypes
oil.loc[:,'Dubai(BPD)':'WTI(BPD)']=oil.loc[:,'Dubai(BPD)':'WTI(BPD)'].astype('float64')
oil
```

```python
## 환율
start_date = '2019-01-01'
end_date = '2021-12-31'
data = yf.download(['USDKRW=X','JPYKRW=X'],start=start_date, end=end_date)
exchange = data['Close']
exchange.rename(columns={'JPYKRW=X':'JPY(1YEN)','USDKRW=X':'USD(1$)'},inplace=True)
exchange
```

```python
## 한국 국채 수익률
kr_3year = pd.read_csv('3y.csv')
kr_3year.rename(columns={'kr_3year_yield':'KR_3year_yield(%)'},inplace=True)
pd.to_datetime(kr_3year['Date'])
kr_3year.set_index('Date',inplace=True)

kr_3year.index = kr_3year.index.astype('datetime64[ns]')
kr_3year
```

```python
## 미국 국채 수익률
start_date = '2019-01-01'
end_date = '2021-12-31'

us_10year = fdr.DataReader('US10YT=X', start=start_date, end=end_date)
us_10year.rename(columns={'Close':'US_10year_yeild(%)'},inplace=True)
us_10year_yeild= us_10year['US_10year_yeild(%)']
us_10year_yeild = pd.DataFrame(us_10year_yeild)
us_10year_yeild
```

```python
### stock_index_1: 주가 + 채권 자료 
stock_index_1=stock_index.merge(kr_3year,on='Date',how='left')
stock_index_1
```

```python
## stock_index_2: stock_index_1 + 미국 국채
stock_index_2=stock_index_1.merge(us_10year_yeild,on='Date',how='left')
stock_index_2
```

```python
## stock_index_3: stock_index_2 + 금 현물
stock_index_3=stock_index_2.merge(Gold_set,on='Date',how='left')
stock_index_3
```

```python
#stock_index_4 : stock_index_3 + 유가
stock_index_4=stock_index_3.merge(oil,on='Date',how='left')
stock_index_4
```

# 데이터 스케일링
* 종류
    - standard
    - minmax
    - log
    - log1p
    
## 주의사항
* log, log1p가 음수에 대해서 결측치로 반환한다.!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
* 결측치: 2020-4-20 WTI(BPD) -37.64 너무 커서 log1p도 결측치 반환한다.
* log,log1p에 대해서 결측치에 interpolate를 적용했다.... 되려나???????????

```python
# 전체 데이터에 대한 스케일링(standard, minmax,log,log1)
# standard_df: standard_scaler
# minmax_df: minmax_scaler
# log_df: log변환
# log1_df: log(1+x) 변환



from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np


log_df = np.log(stock_index_6)
log_df.fillna(0,inplace=True)

log1_df = np.log1p(stock_index_6)
log1_df.fillna(0,inplace=True)

standard = StandardScaler()
minmax = MinMaxScaler()

standard_scale = standard.fit_transform(stock_index_6)
minmax_scale = minmax.fit_transform(stock_index_6)

standard_df = pd.DataFrame(standard_scale,columns=stock_index_6.columns,index=stock_index_6.index)
minmax_df = pd.DataFrame(minmax_scale,columns=stock_index_6.columns,index=stock_index_6.index)



# 스케일링 데이터 시각화
standard_df.plot()
plt.title('standard scale')

minmax_df.plot()
plt.title('minmax scale')

log_df.plot()
plt.title('log scale')

log1_df.plot()
plt.title('log1 scale')
```

## raw data: Lasso, Ridge, random forest

```python
# feature selection->raw data: Lasso, Ridge, random forest
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import seaborn as sns

# Lasso를 이용한 모형의 상관계수
lasso = Lasso()
lasso.fit(features,target)
coeff = pd.Series(data=lasso.coef_,index=features.columns)
coeff = coeff.sort_values(ascending=True)
sns.barplot(x=coeff.values,y=coeff.index)
plt.show()



# Ridge를 이용한 모형의 상관계수
ridge = Ridge()
ridge.fit(features,target)
coeff = pd.Series(data=ridge.coef_,index=features.columns)
coeff = coeff.sort_values(ascending=True)
sns.barplot(x=coeff.values,y=coeff.index)
plt.show()


# RandomForest를 이용한 모형의 중요도
rf_rg=RandomForestRegressor()
# X_train.X_test,y_train,y_test = train_test_split(features,target,test_size=0.3,random_state=156)

rf_rg.fit(features,target)

sns.barplot(x=rf_rg.feature_importances_,y=features.columns)
plt.show()
```

* 원자재: 유종 1개, + 금
