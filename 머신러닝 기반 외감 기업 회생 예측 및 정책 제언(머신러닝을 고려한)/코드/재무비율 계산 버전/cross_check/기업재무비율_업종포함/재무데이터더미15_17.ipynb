```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/LG/Desktop/금융 빅데이터/프로젝트2/코드/재무비율 계산 버전/최종rawdata/12_18외감.csv',encoding='cp949')



for i in np.arange(len(df.columns)): 
    ### 0에서부터 df의 컬럼 개수에 해당하는 정수로 구성된 1차원 행렬 구성
    ### i에 0부터 df의 컬럼 개수까지 정수를 넣어서 반복한다.
    name = re.sub(pattern = '\([^)]*\)',repl='',string=df.columns[i])
    ### 각각의 컬럼명마다 괄호 안에 있는 모든 문자를 제거하여 name이라는 변수에 저장
    df.rename(columns={df.columns[i]:name},inplace=True)
    ### 컬럼명을 name에 저장된 값으로 바꾼다.

df.columns =df.columns.str.replace('.1','')
df.columns =df.columns.str.replace('.2','')
df.columns =df.columns.str.replace('.3','')
df.columns =df.columns.str.replace('.4','')
df.columns =df.columns.str.replace('.5','')
df.columns =df.columns.str.replace('.6','')

df.columns = df.columns + df.loc[0,:]
df.drop([0,1],axis=0,inplace=True)
df.drop('항목년도',axis=1,inplace=True)
df.rename(columns={np.nan:'구분'},inplace=True)
df.set_index('구분',inplace=True)
df = df.astype('float64')    

df
com15 = pd.read_csv('./더미화/정상기간/15_17이자(업종더미).csv',encoding='cp949')
com15.set_index('구분',inplace=True)
```

```python
df
```

```python
df15 = df[['자산2017','매출액2017']]
df15
com15
com_fin15 = com15.join(df15)
com_fin15
```

```python
a = com_fin15['자산2017'].quantile(0.25)
b = com_fin15['자산2017'].quantile(0.75)

com_fin15.reset_index(inplace=True)

for i in com_fin15.index:
    if com_fin15['자산2017'][i].astype('float64') < a:
        com_fin15['자산2017'][i] = int(0)
    
    elif b <= com_fin15['자산2017'][i].astype('float64'):
        com_fin15['자산2017'][i] = int(2)
    
    else:
        com_fin15['자산2017'][i] = int(1)

com_fin15['자산2017']
```

```python
a = com_fin15['매출액2017'].quantile(0.25)
b = com_fin15['매출액2017'].quantile(0.75)

com_fin15.reset_index(inplace=True)

for i in com_fin15.index:
    if com_fin15['매출액2017'][i].astype('float64') < a:
        com_fin15['매출액2017'][i] = int(0)
    
    elif b <= com_fin15['매출액2017'][i].astype('float64'):
        com_fin15['매출액2017'][i] = int(2)
    
    else:
        com_fin15['매출액2017'][i] = int(1)

com_fin15['매출액2017']
```

```python
com_fin15.drop('index',axis=1,inplace=True)
```

```python
m = com_fin15['자산2017']
s = com_fin15['매출액2017']

com_fin15.drop(['자산2017','매출액2017'],axis=1,inplace=True)
com_fin15.insert(32,column = '자산',value = m)
com_fin15.insert(32,column = '매출',value = s)
com_fin15
```

```python
com_fin15[['자산','매출']] = com_fin15[['자산','매출']].astype(int)
com_fin15
```
